```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import auc
import seaborn as sns
```

```python
PATH = "/home/leonardo/Documents/GANomaly/anomaly-flow/plots"
```

```python

roc_curves = [
    "NF-ToN-IoT-v2-DDoS-downsample",
    "Cross_Evaluation_NF-ToN-IoT-v2-DDoS-downsample-NF-CSE-CIC-IDS2018-v2-DDoS-downsample",
    "Cross_Evaluation_NF-ToN-IoT-v2-DDoS-downsample-NF-BoT-IoT-v2-DDoS-downsample"
]

labels = [
    "ToN-IoT Local Training/Test",
    "ToN-IoT/CIC-IDS Cross Evaluation", 
    "ToN-IoT/BoT-IoT Cross Evaluation", 
]

plt.figure(figsize=(16, 12))
sns.set(style="darkgrid")
sns.set_palette("bright")
sns.set(font_scale=1.5)

for i, curve in enumerate(roc_curves):
    fpr = np.load(f"{PATH}/{curve}/{curve}_fpr.npy")
    tpr = np.load(f"{PATH}/{curve}/{curve}_tpr.npy")

    roc_auc = auc(fpr, tpr)

    # Plot the ROC curve
    plt.plot(fpr, tpr, lw=2, label=f'{labels[i]} (AUC = {roc_auc:.3f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve Model Trained on ToN-IoT Dataset')
    plt.legend(loc='lower right')

# Show the plot
plt.show()
```
