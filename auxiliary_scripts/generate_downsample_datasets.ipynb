```python
import pandas as pd
```

```python
DOWNSAMPLE_DATASET = "NF-UNSW-NB15-v2"
```

```python

df = pd.read_csv(f"../datasets/{DOWNSAMPLE_DATASET}.csv.gz")
```

```python
df.columns
```

```python
# Group by 'class'
groups = df.groupby('Attack')
```

```python
print(groups)
```

```python
groups.all()
```

```python
# Define the resampling proportion
resample_proportion = 0.2

# Resample within each group while maintaining the class proportions
resampled_df = groups.apply(lambda group: group.sample(frac=resample_proportion, replace=True)).reset_index(drop=True)
```

```python
resampled_df['Label'].value_counts()
```

```python
file_path = f"./datasets/{DOWNSAMPLE_DATASET}-downsample.csv"
resampled_df.to_csv(file_path, index=False)
```
